```python

from sqlalchemy import select
from sqlalchemy.orm import Session, lazyload, joinedload
from sqlalchemy import create_engine
from collections import defaultdict
import json

from sqlalchemy import and_

import pandas as pd

import toml
import datetime

import matplotlib.pyplot as plt

from ss13tools.settings import make_engine
from ss13tools.model import Round, Feedback

engine = make_engine("settings.toml")
```

```python
import matplotlib.dates as pltd
```

```python
session = Session(engine)
```

```python
cere_rounds = session.query(Round).filter(Round.map_name=='cerestation', Round.start_datetime >= datetime.datetime(2023, 1, 17)).order_by(Round.start_datetime).options(lazyload(Round.feedbacks), joinedload(Round.populations)).all()
```

```python
table = pd.read_sql_query(
    session.query(Round).filter(Round.map_name=='cerestation', Round.start_datetime >= datetime.datetime(2023, 1, 17)).order_by(Round.start_datetime).options(lazyload(Round.feedbacks), joinedload(Round.populations)).statement,
    engine.connect()
)
```
