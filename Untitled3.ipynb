```python
engine = make_engine("settings.toml")
session = Session(engine)
```

```python
query = select(Feedback).join(Round).filter(
        Feedback.key_name == 'events',
        Round.id.in_([41359, 41634, 41652, 41672]))
```

```python
events = defaultdict(set)
for feedback in session.scalars(query).all():
    for severity, kinds in feedback.items():
        events[feedback.round.id].update(list(kinds.keys()))
```
