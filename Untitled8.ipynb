```python
%run shared_imports.py
from decimal import Decimal
from dataclasses import dataclass
from sqlalchemy.orm.query import Query
```

```python
engine = make_engine("settings.toml")
session = Session(engine)
```

```python
query = session.query(Feedback.datetime, Feedback.json).join(Round).filter(
    Feedback.key_name == 'RND Production List',
    Round.start_datetime >= datetime(2024, 3, 30),
    Round.start_datetime < datetime(2024, 9, 30))

rnd_prod_lists = pd.read_sql_query(query.statement, session.connection())
```

```python
rnd_prod_lists
```

```python
parts_kits_names = set()

def apply_parts_kits_count(text):
    l = text['data']['/list']
    result = list()
    for k, v in l.items():
        if 'Parts Kit' in k:
            name = k.replace(' Parts Kit', '')
            parts_kits_names.add(name)
            result.append((name, v))
    if len(result):
        keys, values = zip(*result)
        series = pd.Series(values, index=keys)
        # print(series)
        return series
    else:
        return pd.Series()

def get_parts_kits_count(df):
    return pd.concat([df, df['json'].apply(apply_parts_kits_count)], axis=1).drop(['json'], axis=1)
```

```python
parts_kits = get_parts_kits_count(rnd_prod_lists)
```

```python
sns.set_theme()
```

```python
parts_kits
```

```python
lessers = set(['Arc Revolver', 'u-ION Silencer', 'Energy Crossbow', 'Plasma Pistol'])

def set_date(df):
    df['month'] = df.datetime.dt.month
    df['day'] = df.datetime.dt.day
    df['year'] = df.datetime.dt.year
    return df

datumed_parts_kits = set_date(parts_kits)
data = dict()
for kit in sorted(parts_kits_names - lessers):
    datum = datumed_parts_kits.groupby(['year', 'month', 'day'], as_index=False)[kit].sum(numeric_only=True)
    datum['dt'] = pd.to_datetime(datum.year.astype(str) + datum.month.astype(str) + datum.day.astype(str), format='%Y%m%d')
    datum.drop(['month', 'day', 'year'], inplace=True, axis=1)
    data[kit] = datum
```
