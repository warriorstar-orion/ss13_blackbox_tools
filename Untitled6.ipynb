```python
from pathlib import Path
import math
```

```python
dme = DME.from_file("D:/ExternalRepos/third_party/Paradise/paradise.dme")
space_ruins = Path("D:/ExternalRepos/third_party/Paradise/_maps/map_files/RandomRuins/SpaceRuins/")
lava_ruins = Path("D:/ExternalRepos/third_party/Paradise/_maps/map_files/RandomRuins/LavaRuins/")
```

```python
spaceruin_map_data = dme.subtypesof('/datum/map_template/ruin/space')
lavaruin_map_data = dme.subtypesof('/datum/map_template/ruin/lavaland')
```

```python
engine = make_engine("settings.toml")
session = Session(engine)
```

```python
query = session.query(Feedback.datetime, Feedback.json).join(Round).filter(
        Feedback.key_name == 'ruin_placement',
        Round.start_datetime >= datetime(2024, 8, 5),
        Round.start_datetime < datetime(2024, 10, 5))

raw_prod_ruins = pd.read_sql_query(query.statement, session.connection())
```

```python
spaceruin_typedecls = [dme.typedecl(x) for x in spaceruin_map_data]
lavaruin_typedecls = [dme.typedecl(x) for x in lavaruin_map_data]
```

```python
spaceruin_dmms = [DMM.from_file(space_ruins / td.value('suffix')) for td in spaceruin_typedecls]
lavaruin_dmms = [DMM.from_file(lava_ruins / td.value('suffix')) for td in lavaruin_typedecls if td.value('suffix')]
```

```python
lava_dmm_costs = dict()
space_dmm_costs = dict()
for ruin in spaceruin_dmms:
    space_dmm_costs[ruin.filepath.name] = math.floor(math.sqrt(ruin.extents[0] * ruin.extents[1]))
for ruin in lavaruin_dmms:
    lava_dmm_costs[ruin.filepath.name] = math.floor(math.sqrt(ruin.extents[0] * ruin.extents[1]))
```

```python
normalized = pd.json_normalize(raw_prod_ruins.json)
```

```python
space_totals = list()
lava_totals = list()
for index, row in normalized.iterrows():
    space_row_costs = list()
    lava_row_costs = list()
    for r in row:
        if str(r).endswith(".dmm"):
            if str(r) in space_dmm_costs:
                space_row_costs.append(space_dmm_costs[str(r)])
            elif str(r) in lava_dmm_costs:
                lava_row_costs.append(lava_dmm_costs[str(r)])
    space_totals.append(sum(space_row_costs))
    lava_totals.append(sum(lava_row_costs))
```
