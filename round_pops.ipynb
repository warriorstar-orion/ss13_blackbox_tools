```python
max_rounds_per_mapmonth = defaultdict(lambda: defaultdict(list))
for map_name in ['cerestation', 'Delta', 'MetaStation', 'Cyberiad']:
    rounds = session.query(Round).filter(Round.map_name==map_name, Round.start_datetime >= datetime.datetime(2023, 1, 17)).order_by(Round.start_datetime).options(lazyload(Round.feedbacks), joinedload(Round.populations))
    for round in rounds.all():
        month = round.start_datetime.strftime("%Y-%m")
        max_rounds_per_mapmonth[map_name][month].append(round.highest_player_count)
```

```python
rows = dict()
for mapname, months in max_rounds_per_mapmonth.items():
    row = {month: mean(counts) for month, counts in months.items()}
    rows[mapname] = row
```

```python
all_months = set()
for mapname, months in rows.items():
    all_months.update(months.keys())
```

```python
column_names = ["map"] + sorted(all_months)
print("\t".join(column_names))
for map in sorted(rows.keys()):
    print(f"{map}\t", end='')
    months = rows[map]
    for month in sorted(all_months):
        if month in months:
            print(f"{months[month]:.0f}\t", end='')
        else:
            print("0\t", end='')
    print()

    
```

```python
roundstart_manifests = defaultdict(lambda: defaultdict(list))
for map_name in ['cerestation', 'Delta', 'MetaStation', 'Cyberiad']:
    rounds = session.query(Round).filter(Round.map_name==map_name, Round.start_datetime >= datetime.datetime(2023, 1, 17)).order_by(Round.start_datetime).options(lazyload(Round.feedbacks), joinedload(Round.populations))
    for round in rounds.all():
        month = round.start_datetime.strftime("%Y-%m")
        if not round.has_feedback('manifest'):
            continue
        roundstart_manifests[map_name][month].append(round.roundstart_job_count(with_assts=True))
```

```python
rows = dict()
for mapname, months in roundstart_manifests.items():
    row = {month: mean(counts) for month, counts in months.items()}
    rows[mapname] = row
```

```python
entire_round_manifests = defaultdict(lambda: defaultdict(list))
for map_name in ['cerestation', 'Delta', 'MetaStation', 'Cyberiad']:
    rounds = session.query(Round).filter(Round.map_name==map_name, Round.start_datetime >= datetime.datetime(2023, 1, 17)).order_by(Round.start_datetime).options(lazyload(Round.feedbacks), joinedload(Round.populations))
    for round in rounds.all():
        month = round.start_datetime.strftime("%Y-%m")
        if not round.has_feedback('manifest'):
            continue
        roundstart_count = sum([y.get("roundstart", 0) for x,y in round.feedback('manifest').items()])
        latejoin_count = sum([y.get("latejoin", 0) for x,y in round.feedback('manifest').items()])
        entire_round_manifests[map_name][month].append(roundstart_count + latejoin_count)

rows = dict()
for mapname, months in entire_round_manifests.items():
    row = {month: mean(counts) for month, counts in months.items()}
    rows[mapname] = row

all_months = set()
for mapname, months in rows.items():
    all_months.update(months.keys())

column_names = ["map"] + sorted(all_months)
print("\t".join(column_names))
for map in sorted(rows.keys()):
    print(f"{map}\t", end='')
    months = rows[map]
    for month in sorted(all_months):
        if month in months:
            print(f"{months[month]:.0f}\t", end='')
        else:
            print("0\t", end='')
    print()

    
```

```python
with Session(engine) as session:
    latejoin_manifests = defaultdict(lambda: defaultdict(list))
    for map_name in ['cerestation', 'Delta', 'MetaStation', 'Cyberiad']:
        feedbacks = session.query(Feedback).join(Round).filter(Feedback.key_name=='manifest', Round.map_name==map_name, Round.start_datetime >= datetime.datetime(2023, 1, 17))
        for feedback in feedbacks.all():
            round = feedback.round
            month = round.start_datetime.strftime("%Y-%m")
            latejoin_count = sum([y.get("latejoin", 0) for x,y in feedback.items()])
            latejoin_manifests[map_name][month].append(latejoin_count)
    
    rows = dict()
    for mapname, months in latejoin_manifests.items():
        row = {month: mean(counts) for month, counts in months.items()}
        rows[mapname] = row
    
    all_months = set()
    for mapname, months in rows.items():
        all_months.update(months.keys())
    
    column_names = ["map"] + sorted(all_months)
    print("\t".join(column_names))
    for map in sorted(rows.keys()):
        print(f"{map}\t", end='')
        months = rows[map]
        for month in sorted(all_months):
            if month in months:
                print(f"{months[month]:.0f}\t", end='')
            else:
                print("0\t", end='')
        print()
    
        
```

```python
with Session(engine) as session:
    mapping_votecounts = defaultdict(lambda: defaultdict(list))
    feedbacks = session.query(Feedback).join(Round).filter(Feedback.key_name=='votes', Round.start_datetime >= datetime.datetime(2023, 1, 17))
    for feedback in feedbacks.all():
        month = feedback.round.start_datetime.strftime("%Y-%m")
        if "map" not in feedback.json_data():
            continue
        for mapname, count in feedback.json["data"]["map"].items():
            mapping_votecounts[mapname][month].append(count)
    rows = dict()
    for mapname, months in mapping_votecounts.items():
        row = {month: mean(counts) for month, counts in months.items()}
        rows[mapname] = row

    all_months = set()
    for mapname, months in rows.items():
        all_months.update(months.keys())

    column_names = ["map"] + sorted(all_months)
    print("\t".join(column_names))
    for map in sorted(rows.keys()):
        print(f"{map}\t", end='')
        months = rows[map]
        for month in sorted(all_months):
            if month in months:
                print(f"{months[month]:.0f}\t", end='')
            else:
                print("0\t", end='')
        print()
```
