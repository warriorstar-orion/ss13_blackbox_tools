```python
from model import Round
from sqlalchemy import select
from sqlalchemy.orm import Session
from sqlalchemy import create_engine

import toml

config = toml.load(open('config.toml'))
connection_string = config['database']['prod_connection_string']
engine = create_engine(connection_string)
```

```python
from collections import defaultdict

r_count = 0
total_count = 0

orders = defaultdict(int)

items = defaultdict(int)
with Session(engine) as session:
    for r in session.execute(select(Round).order_by(Round.id)).scalars():
        if r.has_feedback_stat("cargo_exports"):
            round_orders = r.get_feedback_stat("cargo_exports")
            for k, v in round_orders.items():
                orders[k] += v
            r_count += 1

```
